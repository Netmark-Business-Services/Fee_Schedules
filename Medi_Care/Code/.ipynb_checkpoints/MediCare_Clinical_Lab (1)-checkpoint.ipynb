```python
import pandas as pd
import re
file_path=r"../Input_Files/CLF 2024 Q2V1/CLFS 2024 Q2V1.xlsx"
df = pd.read_excel(file_path)
```

```python
display(df)
```

```python
header_row = df[df.iloc[:, 0].str.contains('YEAR', na=False)].index[0] + 1

# Read the data again using the correct header row
df_clean = pd.read_excel(file_path, header=header_row)
```

```python
display(df_clean)
```

```python
columns = df_clean.columns
```

```python
rate_columns = [col for col in df_clean.columns if col.startswith('RATE')]

# Select the required columns including the first 'RATE' column
df_selected = df_clean[['HCPCS', 'EFF_DATE', rate_columns[0]]]
```

```python
display(df_selected)
```

```python
output_path = '../Output_Files/MCare_Clinical_Lab.xlsx'

# Export the DataFrame to an Excel workbook
df_selected.to_excel(output_path, index=False)
```
