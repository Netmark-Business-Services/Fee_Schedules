```python
import pandas as pd
import re
df = pd.read_excel(r"../Input_Files/April_2024_ASP_Pricing_File_032124___Final_File/April 2024 ASP Pricing File 032124.xlsx")
```

```python
display(df)
```

```python
header_row = df[df.iloc[:, 0].str.contains('HCPCS Code', na=False)].index[0] + 1

# Read the data again using the correct header row
df_clean = pd.read_excel(r"../Input_Files/April_2024_ASP_Pricing_File_032124___Final_File/April 2024 ASP Pricing File 032124.xlsx", sheet_name='APR_24_ASP_byHCPCS', header=header_row)
```

```python
display(df_clean)
```

```python
df_clean.columns
```

```python

# Select only the relevant columns
df_extracted = df_clean[['HCPCS Code', 'Payment Limit']]

#Rename the column names
df_extracted = df_extracted.rename(columns={
    'HCPCS Code': 'SVC_CD',
    'Payment Limit': 'FEE'
})


# Add the additional required columns with default values as they are not present in the file
df_extracted['MODIFIER_1'] = ''
df_extracted['MODIFIER_2'] = ''
df_extracted['MODIFIER_3'] = ''
df_extracted['MODIFIER_4'] = ''
df_extracted['AS_OF_DATE'] = ''
df_extracted['AGE_UNIT'] = ''
df_extracted['FROM_AGE'] = ''
df_extracted['THROUGH_AGE'] = ''
df_extracted['PLACE_OF_SERVICE_CODE'] = ''

# Reorder the columns to match the required format
df_extracted = df_extracted[['SVC_CD', 'MODIFIER_1', 'MODIFIER_2', 'MODIFIER_3', 'MODIFIER_4', 'FEE', 'AS_OF_DATE', 'AGE_UNIT', 'FROM_AGE', 'THROUGH_AGE', 'PLACE_OF_SERVICE_CODE']]
```

```python
display(df_extracted)
```

```python
date_pattern = r"\b\d{6}\b"
    
# Search for the date pattern in the file path
match = re.search(date_pattern, r"../Input_Files/April_2024_ASP_Pricing_File_032124___Final_File/April 2024 ASP Pricing File 032124.xlsx")
    
# Extract and format the date if found
extracted_date = match.group()
formatted_date = f"20{extracted_date[-2:]}-{extracted_date[-4:-2]}-{extracted_date[:-4]}"
```

```python
# Save the extracted data to a new Excel file
output_file_path = f'../Output_Files/MCare_ASP_{formatted_date}.xlsx'
df_extracted.to_excel(output_file_path, index=False)
```
